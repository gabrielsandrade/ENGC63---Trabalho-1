```python
import numpy as np
import matplotlib.pyplot as plt
```

### Função de plotagem de gráficos

```python
N = 20
n = np.arange(0, N, 1)
axis = np.arange(0, 1, 1/N)

def plot_graph(data, title, axis=axis, plot_h_line=True):
    plt.figure(figsize=(15, 8), dpi=80)
    max_value = np.amax(data)
    min_value = np.amin(data)
    if plot_h_line:
        plt.axhline(0, color='black')
    
    plt.axvline(0, color='black')
    plt.axhline(max_value, color='green')
    plt.axhline(min_value, color='green')
    plt.title(title)
    plt.plot(axis, data, 'o', color='blue')
    plt.xticks(axis)
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
```

### Sinal

```python
# x1[n] = cos(0, 2πn)
# x2[n] = cos(0, 4πn − π/2)
# x3[n] = cos(0, 8πn + π/5)

x1 = np.cos(0.2 * np.pi * n)
x2 = np.cos(-np.pi/2 + (0.4 * np.pi * n))
x3 = np.cos(np.pi/5 + (0.8 * np.pi * n))
```

```python
plot_graph(x1, 'x1[n]', np.arange(-N/2, N/2, 1))
```

```python
plot_graph(x2, 'x2[n]',np.arange(-N/2, N/2, 1))
```

```python
plot_graph(x3, 'x3[n]', np.arange(-N/2, N/2, 1))
```

```python
x = x1 + x2 + x3
plot_graph(x, 'x = x1[n] + x2[n] + x3[n]', plot_h_line=False)
x
```

## Magnitude normalizada

```python
X = np.fft.fft(x)
magnitude = np.real(X)
plot_graph(magnitude, 'Magnitude normalizada')
```

## Espectro de fase

```python
phase = np.angle(X)
plot_graph(phase, 'phase')
```

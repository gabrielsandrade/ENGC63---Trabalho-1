```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_graph(data, title):
    plt.plot(np.arange(-N, N), 0*np.arange(-N, N))
    plt.axvline(0)
    plt.plot( np.arange(-N, N), data, 'o')
    plt.xticks(np.arange(-N, N, 1.0))
    plt.show()
    
```

```python
N = 10
x1 = np.cos((0.2 * np.pi * np.arange(-N, N)))
plot_graph(x1, 'teste')
```

```python
X = np.fft.fft(x1)
plot_graph(X, '')
```

```python
# x1[n] = cos(0, 2πn)
# x2[n] = cos(0, 4πn − π/2)
# x3[n] = cos(0, 8πn + π/5)

x2 = np.cos(((0.4 * np.pi) - (np.pi/2)) * np.arange(-N, N))
plot_graph(x2, 'teste2')
```

```python
# x1[n] = cos(0, 2πn)
# x2[n] = cos(0, 4πn − π/2)
# x3[n] = cos(0, 8πn + π/5)

x3 = np.cos(0.8 * np.pi + (np.pi/5) * np.arange(-N, N) )
plot_graph(x3, 'teste3')
```
